```python
# Import packages
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
from sklearn.metrics import silhouette_samples, silhouette_score
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Read in CSV and list columns
df = pd.read_csv('medical_clean.csv')
df.columns
```

# Prepare Data Set

```python
# Explore initial variable types and non-null counts
columns_of_interest = ['Population', 'Lat', 'Lng', 'Area', 'ReAdmis']

df[columns_of_interest].info()
```

```python
# Check to see if any complete duplicates exist
df.duplicated().value_counts()
```

```python
# Check value counts for each column of interest to determine if a non-normal value was used as a null value such as a '-' or 'null'.

for col in columns_of_interest:
    print(df[col].value_counts())
```

```python
# Boxplot to review outliers
plt.figure(figsize=(8,4))

df.boxplot(column=['Population', 'Lat', 'Lng'])

plt.title('Boxplot for Columns of Interest')
plt.ylabel('Values')
plt.show()
```

```python
# Encode categorical features with OneHotEncoding
# Define categorical features
# Apply OneHotEncoding
encoded_df = pd.get_dummies(df, columns=['Area'], dtype=int)
encoded_df[['Area_Rural', 'Area_Suburban', 'Area_Urban']].head()
```

```python
# Scale numeric features with StandardScaler
numeric_features = ['Population', 'Lat', 'Lng']
scaler = StandardScaler()
encoded_df[numeric_features] = scaler.fit_transform(encoded_df[numeric_features])
encoded_df[numeric_features].head()
```

```python
# Export prepared data set to CSV
encoded_df.to_csv('Prepared Data Set.csv', index=False)
```

# Hierarchical Cluster Analysis

```python
# Assign selected features to variable X
features = ['Population', 'Lat', 'Lng', 'Area_Rural', 'Area_Suburban', 'Area_Urban']
X=encoded_df[features]
X.head()
```

```python
# Perform linkage using the 'Ward' method (Daityari, n.d.a)
linked = linkage(X, method='ward')
plt.figure(figsize=(10,7))

# Produce the dendrogram (Daityari, n.d.b)
dendrogram(linked, orientation='top', distance_sort='descending', show_leaf_counts=True)
plt.show()
```

```python
# 50 = 8, 56 = 7, 63 = 6, 64 = 4, 80 = 3
# List of potential cutpoints
potential_cutpoints = [50, 56, 63, 64, 80]

# Loop through list of potential cut points, cut cluster, produce value counts and silhouette chart
for cut in potential_cutpoints:
    # Perform the cut using fcluster
    clusters = fcluster(linked, cut, criterion='distance')
    
    # Assign clusters back to original dataframe
    df['Cluster'] = clusters
    
    # Print number of clusters and their distribution of value counts
    print(f'Cut {cut} produced {df["Cluster"].nunique()} clusters:')
    print(df['Cluster'].value_counts())
    print()


    # Retrieve the silhouette values (Scikit-learn, 2024)
    silhouette_vals = silhouette_samples(X, clusters)
    cluster_labels = np.unique(clusters)

    # Plot the silhouette values (Scikit-learn, 2024)
    plt.figure(figsize=(8, 4))
    y_ax_lower, y_ax_upper = 0, 0
    for i, c in enumerate(cluster_labels):
        c_silhouette_vals = silhouette_vals[clusters == c]
        c_silhouette_vals.sort()
        y_ax_upper += len(c_silhouette_vals)
        plt.barh(range(y_ax_lower, y_ax_upper), c_silhouette_vals, height=1.0, edgecolor='none')
        y_ax_lower += len(c_silhouette_vals)

    # Add a line representing the silhouette score (the average silhouette value) to the chart (Scikit-learn, 2024)
    plt.axvline(silhouette_score(X, clusters), color="red", linestyle="--")
    plt.xlabel('Silhouette coefficient values')
    plt.ylabel('Cluster labels')
    plt.title(f'Silhouette plot for hierarchical clustering with cut at {cut}')
    plt.show()

```

```python
# Perform cut again with optimal cut point (56)
clusters = fcluster(linked, 56, criterion='distance')
    
# Assign clusters back to original dataframe
df['Cluster'] = clusters
print(f'Cut 56 produced {df["Cluster"].nunique()} clusters:')
print(df['Cluster'].value_counts())
```

```python
# Replace ReAdmis values with 1/0 so we can calculate the rate
replacement_dict = {
    'Yes': '1',
    'No': '0'
}
df['ReAdmis'] = df['ReAdmis'].replace(replacement_dict)
df['ReAdmis'] = df['ReAdmis'].astype(int)
df['ReAdmis'].value_counts()
```

```python
# Calculate the ReAdmis rate per cluster
df.groupby('Cluster')['ReAdmis'].mean()
```

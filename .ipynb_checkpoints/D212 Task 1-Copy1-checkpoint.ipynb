```python
# Import packages
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import fcluster
```

```python
# Read in CSV and list columns
df = pd.read_csv('medical_clean.csv')
df.columns
```

```python
# Encode categorical features with OneHotEncoding
# Define categorical features
# Apply OneHotEncoding
encoded_df = pd.get_dummies(df, columns=['Area'])
encoded_df.head()
```

```python
# Scale numeric features with StandardScaler
numeric_features = ['Population', 'Lat', 'Lng']
scaler = StandardScaler()
encoded_df[numeric_features] = scaler.fit_transform(encoded_df[numeric_features])
encoded_df[numeric_features].head()
```

```python
encoded_df.columns
```

```python
features = ['Population', 'Lat', 'Lng', 'Area_Rural', 'Area_Suburban', 'Area_Urban']
X=encoded_df[features]
X.head()
```

```python
linked = linkage(X, method='ward')
plt.figure(figsize=(10,7))
dendrogram(linked, orientation='top', distance_sort='descending', show_leaf_counts=True)
plt.show()
```

```python
clusters = fcluster(linked, 50, criterion='distance')
encoded_df['Cluster'] = clusters
encoded_df.head()
```

```python
encoded_df['Cluster'].value_counts()
```

```python
replacement_dict = {
    'Yes': '1',
    'No': '0'
}
encoded_df['ReAdmis'] = encoded_df['ReAdmis'].replace(replacement_dict)
encoded_df['ReAdmis'] = encoded_df['ReAdmis'].astype(int)
encoded_df['ReAdmis'].value_counts()
```

```python
encoded_df.groupby('Cluster')['ReAdmis'].mean()
```
